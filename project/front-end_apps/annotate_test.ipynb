# Manual Tests for annotate.py

Run the cell below to set up the environment for manual testing. The cell provided is all that is needed to set up the enviornment.

Each Test will have instructions and acceptance criteria for whether the test passed.

```python
import annotate
import cv2
import os
import numpy as np

temp_path = "test_img.png"


def get_img():
    if not os.path.isfile(temp_path):
        from matplotlib import pyplot as plt

        plt.plot([1, 2, 3], [1, 2, 3])
        plt.savefig(temp_path)

    return temp_path


def del_img():
    if os.path.isfile(temp_path):
        os.remove(temp_path)
```

## Test open_from_camera

### Test open camera: 
1. Run the cell below. 
2. In the window that shows your camera press SPACEBAR.

Acceptance criteria: 

1. The camera view will be displayed in a window

## Test handle_annotation_ui

### Test open annotation ui: 

1. Run the cell below. 
2. In the annotation window press ESCAPE. 
3. This should test the window opens and no annotations were created.

Acceptance Criteria: 

1. An image of a simple line graph in the annotation window will be seen
2. "success" will be printed to the terminal if everything works

```python
img = cv2.imread(get_img())
annotations = annotate.handle_annotation_ui(img)

assert annotations.annotations == []
print("success")
del_img()
```

### Test successful bounding box: 
1. Run the cell below.
2. Left click drag on the image to create a box
3. Repeat the last step
3. Press CTRL + RIGHT CLICK while in the window
5. Press ESCAPE

Acceptance Criteria:
1. The terminal will print out coordinates of two bounding boxes
2. "success" will be printed to the terminal if everything works

```python
img = cv2.imread(get_img())
annotations = annotate.handle_annotation_ui(img)

assert len(annotations.annotations) == 2
print("success")
del_img()
```
